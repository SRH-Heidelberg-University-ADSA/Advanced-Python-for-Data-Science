## 1. **Software Design Matters**

Many Python programs begin as small scripts. As complexity increases, these scripts evolve into larger systems. Without proper design, software becomes difficult to maintain and extend.

Good software design improves:

- readability
- maintainability
- extensibility
- testability
- collaboration among developers

### Example: poorly designed code

```python
def process(data):
    result = []
    for i in range(len(data)):
        if data[i] > 10:
            result.append(data[i] * 2 + 5)
    return result
```

### Problems

- unclear variable names
- unnecessary indexing
- logic is tightly coupled
- difficult to test individual steps


### Improved Version Using Modular Design

```python
def filter_values(values, threshold):
    return [v for v in values if v > threshold]


def transform_values(values):
    return [v * 2 + 5 for v in values]


def process(values, threshold=10):
    filtered = filter_values(values, threshold)
    return transform_values(filtered)
```

### Advantages

- each function performs one task
- functions are reusable
- individual components can be tested independently

## 2. **Clean Code Principles**

Clean code prioritises readability and simplicity. Code should communicate its purpose clearly.


### 2.1  Meaningful Naming

**Bad example**:
```python
def f(x):
    return x * 2
```

**Improved example**:

```python
def double_value(value):
    return value * 2
```

---

### 2.2 Avoid Magic Numbers

**Bad example**:

```python
def calculate_price(price):
    return price * 1.19
```

**Improved design**:

```python
VAT_RATE = 0.19

def calculate_price(price):
    return price * (1 + VAT_RATE)
```

---

### 2.3 Small Functions

Large functions often perform multiple tasks.

**Bad example**:

```python
def load_and_process(file_path):

    with open(file_path) as f:
        data = f.readlines()

    cleaned = []
    for line in data:
        cleaned.append(int(line.strip()))

    total = sum(cleaned)

    return total / len(cleaned)
```

**Improved version**:

```python
def load_lines(path):
    with open(path) as file:
        return file.readlines()


def clean_numbers(lines):
    return [int(line.strip()) for line in lines]


def compute_mean(values):
    return sum(values) / len(values)
```

## 4. **PEP 8 Coding Standards**

Python Enhancement Proposal 8 (PEP 8) defines the official style guide for Python code.

Important rules include:

| Rule | Example |
|-----|-----|
| functions use snake_case | ```compute_mean()``` |
| classes use CapWords | ```DataLoader``` |
| constants use uppercase | ```MAX_VALUE``` |
| indentation uses four spaces | standard |

---

### **Non-Compliant Code Example**

```python

 sum=0
 for i in DataList:
  sum=sum+i
 return sum/len(DataList)
```

---

### **Corrected Version**

```python
def calculate_mean(data_list):

    total = 0

        total += value

    return total / len(data_list)
```

## 5. **SOLID Design Principles**

SOLID principles improve maintainability in object-oriented software systems.

### 5.1 **Single Responsibility Principle**
Each class should have **a single responsibility**.

**Bad design**:

```python
class Report:

    def generate(self):
        print("Generating report")

    def save(self, path):

        with open(path, "w") as f:

            f.write("Report")
```

**Improved design**:

```python
class ReportGenerator:

    def generate(self):

        return "Report content"


class ReportSaver:

    def save(self, content, path):

        with open(path, "w") as f:

            f.write(content)
```

---

### 5.2 **Open Closed Principle**
Software entities should be **open for extension** but **closed for modification**.

```python
class Discount:

    def apply(self, price):

        raise NotImplementedError
```

**Example subclasses**:

```python
class StudentDiscount(Discount):

    def apply(self, price):

        return price * 0.8


class SeniorDiscount(Discount):

    def apply(self, price):

        return price * 0.7
```

---

### 5.3 **Dependency Inversion Principle**
High-level modules should depend on **abstractions** rather than **implementations**.

```python
class Database:

    def save(self, data):

        print("Saving to database")
```

Service class:

```python
class UserService:

    def __init__(self, database):

        self.database = database

    def create_user(self, user):

        self.database.save(user)
```

```
write test → run test → implement code → refactor
```

### 6.1. **Pytest Parametrize Testing**

Instead of writing separate tests for each value, you can parametrise in ```test_square.py```:

```python
import pytest
from square import square

@pytest.mark.parametrize(
    "input_value, expected",
    [
        (2, 4),
        (3, 9),
        (0, 0),
        (-4, 16),
        (1.5, 2.25)
    ]
)
def test_square(input_value, expected):
    assert square(input_value) == expected
```
Modify ```test_square.py``` accordingly. 

**Explaination**
- ```@pytest.mark.parametrize``` is a pytest **```decorator```** that allows you to run the same test function multiple times with different input values.
- The first argument ```input_value, expected``` lists the parameter names the test function will receive. For example, 2 is the input_value and 4 is the expected output.
- This approach reduces code duplication and automatically tests multiple cases in one function.

### 6.2. **Mocking Strategies in Unit Testing**

Mocking is a fundamental technique in unit testing that allows developers to isolate the code under test by replacing external dependencies with controlled, simulated objects. The goal is to ensure that the behaviour of a unit can be tested independently of other components, such as databases, APIs, or file systems.

a) **Simple Mocks**

A simple mock replaces a real object with a **dummy object** that returns predefined values. It is often used to simulate the responses of functions or methods.

**Example function in ```my_module.py```**:

```python
def fetch_user(api_client):
    """Fetch a user from the given API client."""
    return api_client.get_user()
```

**Testing using mocks in ```my_module.py``` **

```python
from unittest.mock import Mock


def test_fetch_user():
    # create a mock API client
    mock_api = Mock()
    # define what get_user() should return
    mock_api.get_user.return_value = {"name": "Alice"}
    # call the function with the mock
    result = fetch_user(mock_api)
    # assert that the function returns the mocked value
    assert result["name"] == "Alice"
```

b) **Patch Mocking**

```patch``` allows temporarily replacing a target object with a mock during the execution of a test. This is particularly useful for mocking functions, classes, or modules in other parts of the system.

**Example function in ```my_patch.py```**:

```python
def read_file(path):
    """
    Reads the content of a file at the given path and returns it as a string.
    """
    with open(path, "r") as f:
        return f.read()
```
**Testing using patch in ```test_patch.py```**:

```python
from unittest.mock import mock_open, patch
from my_patch import read_file

def test_read_file():
    # Sample content we want the file to "contain"
    mock_content = "Hello, world!"
    
    # patch 'builtins.open' with a mock that simulates file reading
    with patch("builtins.open", mock_open(read_data=mock_content)) as mocked_file:
        result = read_file("file.txt")  # path is irrelevant; the file is mocked
        # assert that the function returns the mocked content
        assert result == mock_content

        # optional: check that open was called with the correct path and mode
        mocked_file.assert_called_once_with("file.txt", "r")
```

d) **Assertion on Mocks**

Mocks can track how they were called, enabling verification of interactions with external dependencies.

```python
def test_fetch_user_return_value():

    Test that fetch_user returns the expected mocked value.

    mock_api = Mock()

    assert result["name"] == "Alice"

    mock_api.get_user.assert_called_once()

def test_fetch_user_side_effect():

    mock_api.get_user.side_effect = [{"name": "Alice"}, {"name": "Bob"}]

    # first call returns Alice

    assert result1["name"] == "Alice"

    # second call returns Bob
    result2 = fetch_user(mock_api)
```

e) **Mocking Classes and Objects**

Mocks can replace entire classes, allowing instantiation of mock objects instead of real ones. This is useful when testing components that depend on complex objects.

**Example function in ```my_mocks.py```**:

```python
class Database:
    """
    Simulates a database class with a save method.
    """
    def save(self, data):
        # Imagine this writes data to a real database
        print(f"Saving {data} to the database")
        return True

def store_data(db, data):
    """
    Function that saves data using the provided database object.
    """
    return db.save(data)
```
**Testing using patch in ```test_mocks.py```**:

```python
from unittest.mock import patch
from my_mocks import store_data

def test_store_data_mocked_class():
    """
    Test store_data by mocking the Database class entirely.
    """
    # Patch 'Database' in the 'my_mocks' module
    with patch("my_mocks.Database") as MockDatabase:
        # Create a mock instance (this replaces the real Database instance)
        db_instance = MockDatabase()

        # Define the return value of the save() method
        db_instance.save.return_value = True

        # Call the function with the mock instance
        result = store_data(db_instance, {"id": 1, "name": "Alice"})

        # Assert that the function returns True
        assert result is True

        # Verify that save() was called exactly once with the correct argument
        db_instance.save.assert_called_once_with({"id": 1, "name": "Alice"})
```

## 7. **Documentation standards**

Clear documentation improves usability and maintainability.

### a) Docstrings
>```Docstrings``` are the standard way to document Python code. They describe functions, classes, and modules, and can be used by IDEs and tools like Sphinx.

**Guidelines**
- Use triple quotes """ for docstrings.

- Include description, arguments, return value, and exceptions.

-Follow PEP  conventions.

**Function Example**
```python
def square(x):
    """
    Compute the square of a number.

    Args:
        x (int | float): The number to be squared.

    Returns:
        int | float: The square of the input number.

    Raises:
        TypeError: If x is not an int or float.
    """
    if not isinstance(x, (int, float)):
        raise TypeError("input must be numeric")
    return x * x
```
